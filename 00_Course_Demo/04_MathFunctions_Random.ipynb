```python
import numpy as np
```

```python
np.random.seed(42)
vector_nums = np.random.normal(loc = 2, scale = 1, size = 10)

print(vector_nums)
```

```python
np.sin(vector_nums)
```

```python
np.exp(vector_nums)
```

```python
np.log10(vector_nums)
```

```python
np.mean(vector_nums)
```

```python
np.var(vector_nums)
```

```python
np.quantile(vector_nums, [0.25, 0.5, 0.75])
```
